```python
diabetes_numerical=diabetes[numerical]
diabetes_numerical
```

```python
x=diabetes_numerical[['Glucose', 'DiabetesPedigreeFunction']]
x
```

```python
y=diabetes_numerical['Outcome']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
```

```python
from collections import Counter
Counter(y_train)
```

# This is Balanced data

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_prediction=dt.predict(x_test)
y_prediction
```

```python
dt.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_prediction)
cm
```

```python
sns.heatmap(cm,annot=True,fmt='.2f',xticklabels=['no','yes'],yticklabels=['no','yes'])
plt.xlabel('Predicted_class')
plt.ylabel('Actual_class')
```
